Crea un programa que muestre los primeros n números de la serie Fibonacci.

```python
n = int(input("Cantidad: "))
a = 0
b = 1
lista = list(range(n))

for i in lista:
    print(a)
    c = a + b
    a = b
    b = c
```

Escribe un programa que determine si un número es primo.

```python
num = int(input("Ingrese un número: "))
contador = 0

lista = list(range(1,num+1))

for i in lista:
    if num % i == 0:
        contador += 1

if contador == 2:
    print("Es primo")
else:
    print("No es primo")
```

```python
num = int(input("Ingrese un número: "))
contador = 0

for i in range(num):
    if num % (i+1) == 0:
        contador += 1

if contador == 2:
    print("Es primo")
else:
    print("No es primo")
```

```python
num = int(input("Ingrese un número: "))
contador = 0

for i in range(1,num+1,1):
    if num % i == 0:
        contador += 1

if contador == 2:
    print("Es primo")
else:
    print("No es primo")
```
