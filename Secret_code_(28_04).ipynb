```python
escuela = {}

print(" DESCRIBE LA ESCUELA ")

escuela['nombre'] = input("1. ¿Cómo se llama la escuela? ")
escuela['p_principal'] = input("2. Describe la pared frontal y sus dibujos/figuras: ")
escuela['p_trasera'] = input("3. Describe la pared trasera y sus dibujos/figuras: ")
escuela['p_izquierda'] = input("4. Describe la pared izquierda y sus dibujos/figuras: ")
escuela['p_derecha'] = input("5. Describe la pared derecha y sus dibujos/figuras: ")
escuela['techo'] = input("6. ¿De qué color o forma es el techo? ")
escuela['puerta'] = input("7. ¿Cómo es la puerta y sus adornos? ")
escuela['ventanas'] = input("8. ¿Cuántas ventanas tiene y cómo se ven? ")
escuela['baño'] = input("9. Describe el área de los baños: ")
escuela['jardin'] = input("10. Describe detalladamente el jardín: ")
escuela['patio'] = input("11. Describe el patio de juegos y sus atracciones: ")

print("RECORRIDO NARRATIVO POR MI ESCUELA")

descripcion_total = (
    f"Bienvenidos a la escuela **{escuela['nombre']}**. Al detenernos frente a ella, observamos una fachada "
    f"donde {escuela['p_principal']}. Si caminamos hacia la parte posterior, encontramos que la pared trasera "
    f"{escuela['p_trasera']}, mientras que en los costados, la pared izquierda {escuela['p_izquierda']} "
    f"y la pared derecha {escuela['p_derecha']}. Elevando la vista, se aprecia un techo {escuela['techo']} "
    f"que protege todo el lugar. La entrada nos invita a pasar con una puerta {escuela['puerta']} "
    f"y el brillo de sus {escuela['ventanas']}. En su interior, las instalaciones cuentan con un sector de "
    f"baños donde {escuela['baño']}. Para el descanso y la recreación, la escuela ofrece un jardín "
    f"que {escuela['jardin']} y un fantástico patio de juegos donde {escuela['patio']}."
)

print(descripcion_total)
```

```python

colegio = {}

print(" --- 1. ARQUITECTURA (FACHADA Y ESTRUCTURA) --- ")
colegio['nombre'] = input("Nombre del colegio: ")
colegio['colores_fachada'] = input("¿Qué colores tiene la fachada?: ")

# Paredes
tiene_paredes = input("¿Se pueden contar las paredes principales? (si/no): ").lower()
if tiene_paredes == "si":
    colegio['paredes'] = input("   -> ¿Cuántas paredes tiene?: ")
    colegio['mat_paredes'] = input("   -> ¿De qué material son?: ")
else:
    colegio['paredes'] = "varias"
    colegio['mat_paredes'] = "material no especificado"

# Ventanas
tiene_ventanas = input("¿El colegio tiene ventanas? (si/no): ").lower()
if tiene_ventanas == "si":
    colegio['ventanas'] = input("   -> ¿Cuántas ventanas tiene?: ")
    colegio['mat_ventanas'] = input("   -> ¿De qué material son?: ")
else:
    colegio['ventanas'] = "0"
    colegio['mat_ventanas'] = "N/A"

# Puertas
tiene_puertas = input("¿El colegio tiene puertas? (si/no): ").lower()
if tiene_puertas == "si":
    colegio['puertas'] = input("   -> ¿Cuántas puertas tiene?: ")
    colegio['color_puertas'] = input("   -> ¿De qué color son?: ")
    colegio['mat_puertas'] = input("   -> ¿De qué material son?: ")
else:
    colegio['puertas'] = "0"
    colegio['color_puertas'] = "N/A"
    colegio['mat_puertas'] = "N/A"

print(" INSTALACIONES Y ENTORNO ")

# Canchas
if input("¿Tiene canchas deportivas? (si/no): ").lower() == "si":
    colegio['canchas'] = input("   -> ¿De qué material son y cuántas hay?: ")
else:
    colegio['canchas'] = "No cuenta con áreas deportivas"

# Huerta
if input("¿Tiene huerta escolar? (si/no): ").lower() == "si":
    colegio['huerta'] = input("   -> ¿Qué siembran allí?: ")
else:
    colegio['huerta'] = "No tiene zona de siembra"

# Juegos
if input("¿Tiene zona de juegos? (si/no): ").lower() == "si":
    colegio['juegos'] = input("   -> Describe los juegos (columpios, etc): ")
else:
    colegio['juegos'] = "No tiene zona de juegos"

# Comedor
if input("¿Tiene servicio de comedor? (si/no): ").lower() == "si":
    colegio['comedor'] = input("   -> ¿Cómo es el servicio y qué tal es la comida?: ")
else:
    colegio['comedor'] = "No cuenta con servicio de comedor"

print("\n --- 3. INTERIOR Y ACADEMIA --- ")
colegio['piso'] = input("¿Cómo es el piso de los salones?: ")
colegio['pupitres'] = input("¿Cuántos pupitres hay por salón?: ")
colegio['techo'] = input("¿De qué es el techo?: ")
colegio['grados'] = input("¿Qué grados de escolaridad ofrece?: ")
colegio['estudiantes'] = input("¿Cuántos estudiantes hay?: ")
colegio['profesores'] = input("¿Cuántos profesores hay?: ")



# Párrafo 1: Lo físico
desc_fisica = (
    f"El colegio {colegio['nombre']} tiene una fachada de colores {colegio['colores_fachada']}. "
    f"Su estructura principal consta de {colegio['paredes']} paredes de {colegio['mat_paredes']}. "
)
if tiene_ventanas == "si":
    desc_fisica = f"Cuenta con {colegio['ventanas']} ventanas de {colegio['mat_ventanas']}. "
if tiene_puertas == "si":
    desc_fisica = f"Tiene {colegio['puertas']} puertas de {colegio['mat_puertas']} color {colegio['color_puertas']}."

# Párrafo 2: Instalaciones
desc_inst = (
    f"nSobre sus instalaciones, se destaca que: "
    f"Deporte: {colegio['canchas']}."
    f"Recreación: {colegio['juegos']}."
    f"Agricultura: {colegio['huerta']}."
    f"Alimentación: {colegio['comedor']}."
)

# Párrafo 3: Salones y Comunidad
desc_aca = (
    f"En el interior, los salones tienen piso de {colegio['piso']} y {colegio['pupitres']} pupitres. "
    f"El techo es de {colegio['techo']}. Actualmente, la institución ofrece grados {colegio['grados']} "
    f"para un total de {colegio['estudiantes']} alumnos y {colegio['profesores']} docentes."
)

print(desc_fisica)
print(desc_inst)
print(desc_aca)
```
