##1.  PARA - FOR
* Se emplea cuando se sabe cuantas veces se repira la accióm **ojo** El conteo siempre inicia desde **0**

Sintaxis general



```
for variable (variable) in (algo):
  La acción que se va a repetir.

# Forma2
for variable in range(cantidad de veces):
  La acción que se va a repetir.

#Forma3
for variable in range(VI,VF+1,P):
 La acción que se va a repetir.

```
Donde:
* VI:Valor inicial (donde inicio).
* VF:Valor final (donde finalizó).
* P:Paso. Cada cuánto cuento


Sintaxis Para en java

```
for (int variable=VI;variable=VF;
variable=variable+paso){
  La acción que se va a repetir.}
```
