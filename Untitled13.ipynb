```python
n = int(input("Ingrese la cantidad de trabajadores: "))

trabajadores = []
sueldos = []

for i in range(n):
    nombre = input("Nombre del trabajador: ")
    sueldo = float(input("Sueldo: "))

    trabajadores.append(nombre)
    sueldos.append(sueldo)

promedio = sum(sueldos) / n

print("Promedio:", promedio)
print("Trabajadores que ganan más que el promedio:")

for i in range(n):
    if sueldos[i] > promedio:
        print(trabajadores[i])
```

```python
n = int(input("¿Cuántos productos tendrá la tienda?: "))

productos = []
precios = []

for i in range(n):
    producto = input("Nombre del producto: ")
    precio = float(input("Precio: "))

    productos.append(producto)
    precios.append(precio)

print("\nMENÚ DE LA TIENDA")

for i in range(n):
    print(i+1, "-", productos[i], "$", precios[i])
```

```python
cantidad = int(input("Cantidad de números: "))

numeros = []

for i in range(cantidad):
    num = int(input("Ingrese número: "))
    numeros.append(num)

n = int(input("Ingrese el número para saber sus múltiplos: "))

contador = 0

for num in numeros:
    if num % n == 0:
        contador += 1

print("Cantidad de múltiplos:", contador)
```

```python
n = int(input("Cantidad de pacientes: "))

pacientes = []
fechas = []

for i in range(n):
    nombre = input("Nombre del paciente: ")
    vacuna = input("¿Está vacunado? (si/no): ")

    pacientes.append(nombre)

    if vacuna == "si":
        fecha = input("Fecha de vacunación: ")
        fechas.append(fecha)
    else:
        fechas.append("No vacunado")

contador = 0

print("\nPacientes vacunados:")

for i in range(n):
    if fechas[i] != "No vacunado":
        print(pacientes[i], "-", fechas[i])
        contador += 1

print("Total vacunados:", contador)
```
