##Ejercicio 5
Crea un diccionario con nombres de estudiantes y sus notas.
Muestra el promedio de todos los estudiantes y de un estudiante específico, empleando dos funciones.

```python
notas_clase = {
    "Juan": 4.2,
    "Camila": 4.8,
    "Pedro": 3.5
}
def calcular_promedio(diccionario):
    suma_notas = sum(diccionario.values())
    total_estudiantes = len(diccionario)
    return suma_notas / total_estudiantes
def ver_nota_estudiante(nombre):
    return notas_clase[nombre]
print("El promedio del grupo es:", calcular_promedio(notas_clase))
print("La nota de Camila es:", ver_nota_estudiante("Camila"))
```

##Ejercicio 7
Dada la palabra "magia", cuenta cuántas veces aparece cada letra usando un diccionario.

```python
palabra = "magia"
contador = {}
for letra in palabra:
    if letra in contador:
        contador[letra] = contador[letra] + 1
    else:
        contador[letra] = 1
print("Conteo de letras de la palabra magia:", contador)
```

##Ejercicio 8
Dada una lista de elementos recolectados en una expedición, cuenta cuántas veces aparece cada elemento.

```python
elementos = ["gema", "oro", "gema", "madera", "oro", "oro", "gema"]
conteo_elementos = {}
for item in elementos:
    if item in conteo_elementos:
        conteo_elementos[item] += 1
    else:
        conteo_elementos[item] = 1
print("Resultado de la expedición:", conteo_elementos)
```
