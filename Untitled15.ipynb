````python
#1 BATALLA POR EQUIPOS
nombres = ["Ana", "Juan", "Pedro", "Maria", "Luis", "Elena", "Pablo", "Lucia", "Jose", "Rosa"]
notas = [85, 90, 70, 95, 80, 75, 100, 88, 60, 92]

promedio = sum(notas) / 10
print("El promedio de la clase es:", promedio)

print("Estudiantes por encima del promedio ")

for i in range(10):
    if notas[i] > promedio:
        print(nombres[i], "sacó", notas[i])

nota_mayor = max(notas)
posicion = notas.index(nota_mayor)
nombre_mayor = nombres[posicion]

print("El mejor fue", nombre_mayor, "con una nota de:", nota_mayor)
````


```
public class Main {
    public static void main(String[] args) {
        String[] nombres = {"Ana", "Juan", "Pedro", "Maria", "Luis", "Elena", "Pablo", "Lucia", "Jose", "Rosa"};
        int[] notas = {85, 90, 70, 95, 80, 75, 100, 88, 60, 92};


        double suma = 0;
        for (int nota : notas) {
            suma += nota;
        }
        double promedio = suma / notas.length;
        System.out.println("El promedio de la clase es: " + promedio);

        System.out.println("Estudiantes por encima del promedio:");
        for (int i = 0; i < nombres.length; i++) {
            if (notas[i] > promedio) {
                System.out.println(nombres[i] + " sacó " + notas[i]);
            }
        }

        int notaMayor = notas[0];
        int posicion = 0;

        for (int i = 1; i < notas.length; i++) {
            if (notas[i] > notaMayor) {
                notaMayor = notas[i];
                posicion = i;
            }
        }

        String nombreMayor = nombres[posicion];
        System.out.println("\nEl mejor fue " + nombreMayor + " con una nota de: " + notaMayor);
    }
}
```

````python
#2
n = int(input("Ingrese el número de votantes: "))

candidatos = ["abelardo", "cepeda", "paloma", "vicky", "botero"]
votos = [0, 0, 0, 0, 0]

print("Primera vuelta")
print("Opciones:")
for i in range(5):
    print(i+1, "-", candidatos[i])

for i in range(n):
    voto = int(input(f"Votante {i+1}, elija un candidato (1-5): "))
    if 1 <= voto <= 5:
        votos[voto-1] += 1
    else:
        print("Voto inválido")

print("Resultados de la primera vuelta:")
for i in range(5):
    print(candidatos[i], ":", votos[i], "votos")

max1 = 0
max2 = 1

if votos[max2] > votos[max1]:
    max1, max2 = max2, max1

for i in range(2, 5):
    if votos[i] > votos[max1]:
        max2 = max1
        max1 = i
    elif votos[i] > votos[max2]:
        max2 = i

print("Pasan a segunda vuelta:")
print(candidatos[max1])
print(candidatos[max2])

votos2 = [0, 0]

print("Segunda vuelta")
print("1 -", candidatos[max1])
print("2 -", candidatos[max2])

for i in range(n):
    voto = int(input(f"Votante {i+1}, elija (1-2): "))
    if voto == 1:
        votos2[0] += 1
    elif voto == 2:
        votos2[1] += 1
    else:
        print("Voto inválido")

print("Resultados de la segunda vuelta:")
print(candidatos[max1], ":", votos2[0], "votos")
print(candidatos[max2], ":", votos2[1], "votos")

if votos2[0] > votos2[1]:
    print("Ganador:", candidatos[max1])
elif votos2[1] > votos2[0]:
    print("Ganador:", candidatos[max2])
else:
    print("Empate en la segunda vuelta")
````

````python

precios_combos = [20000, 25000, 18000, 24500]
precios_extras = [3000, 7000, 3500, 4000, 2500] # Papas, Jugos, Cerveza, Agua, Gaseosa

print("VENTAS DIARIAS")

total_dia = 0

while True:
    print("\n¿Qué se vendió?")
    print("1-4: Combos | 5: Papas | 6: Jugo | 7: Cerveza | 8: Agua | 9: Gaseosa | 0: Salir y Cerrar Caja")

    opcion = int(input("Seleccione una opción: "))

    if opcion == 0:
        break
    elif 1 <= opcion <= 4:

        total_dia += precios_combos[opcion - 1]
        print(f"Agregado Combo {opcion}")
    elif 5 <= opcion <= 9:

        total_dia += precios_extras[opcion - 5]
        print("Extra agregado correctamente")
    else:
        print("Opción no válida, intenta de nuevo.")

print(f"El total recaudado hoy es: ${total_dia}")

````
