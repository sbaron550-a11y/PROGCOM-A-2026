```python
# Sistema interactivo de asignación de trabajo

class Trabajador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.tareas = []
        self.carga_total = 0

    def asignar_tarea(self, tarea, tiempo):
        self.tareas.append((tarea, tiempo))
        self.carga_total += tiempo


class SistemaAsignacion:
    def __init__(self):
        self.trabajadores = []

    def agregar_trabajadores(self):
        n = int(input("¿Cuántos trabajadores hay?: "))
        for i in range(n):
            nombre = input(f"Nombre del trabajador {i+1}: ")
            self.trabajadores.append(Trabajador(nombre))

    def ingresar_tareas(self):
        tareas = []
        m = int(input("\n¿Cuántas tareas desea ingresar?: "))
        for i in range(m):
            nombre = input(f"Nombre de la tarea {i+1}: ")
            tiempo = int(input(f"Tiempo estimado (horas) de '{nombre}': "))
            tareas.append((nombre, tiempo))
        return tareas

    def asignar_tareas_equilibrado(self, lista_tareas):
        for tarea, tiempo in lista_tareas:
            trabajador = min(self.trabajadores, key=lambda x: x.carga_total)
            trabajador.asignar_tarea(tarea, tiempo)

    def mostrar_resultados(self):
        print("\n RESULTADOS")
        for t in self.trabajadores:
            print(f"\nTrabajador: {t.nombre}")
            print(f"Carga total: {t.carga_total} horas")
            for tarea, tiempo in t.tareas:
                print(f"  - {tarea} ({tiempo}h)")

# EJECUCIÓN DEL PROGRAMA

sistema = SistemaAsignacion()

sistema.agregar_trabajadores()
tareas = sistema.ingresar_tareas()
sistema.asignar_tareas_equilibrado(tareas)
sistema.mostrar_resultados()
```
